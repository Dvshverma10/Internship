```python
# Import required Libraries

import pandas as pd
import time
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
```

```python
# Connect to chromedriver

driver = webdriver.Chrome(r"chromedriver.exe")
```

```python
# FUnction to load URL

def load_url(url):
    
    driver.get(url)
```

```python
# Function to load given url with values

def send_val(key_val, is_xpath, ele_name):
    
    if(is_xpath):
        sel_ele = driver.find_element(By.XPATH,ele_name)
        
    else:
        sel_ele = driver.find_element(By.CLASS_NAME, ele_name)
    
    sel_ele.send_keys(key_val)
```

```python
# Function to perform Submission 

def sub_val(is_xpath, ele_name):
    
    try:
        if(is_xpath):
            sel_ele = driver.find_element(By.XPATH,ele_name)
        
        else:
            sel_ele = driver.find_element(By.CLASS_NAME, ele_name)
            
        sel_ele.click()
        
    except:
        print('')
```

```python
# Function to get List of elements

def get_all_ele(is_xpath, ele_name, is_url):
    
    if(is_xpath):
        sel_ele = driver.find_elements(By.XPATH,ele_name)
        
    else:
        sel_ele = driver.find_elements(By.CLASS_NAME, ele_name) 
        
    in_lst = []
    
    if(is_url):
        
        for i in sel_ele:
            in_lst.append(i.get_attribute('href'))
            
    else:
        
        for i in sel_ele:
            in_lst.append(i.text)
        
    return in_lst  
```
